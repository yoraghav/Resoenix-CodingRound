```python
def wick(i,data):
    if(not(green(i,data))):
        return abs(data.high[i] - data.close[i])
    return abs(data.close[i]-data.low[i])

def body(i,data):
    return abs(data.close[i]-data.open[i])

def green(i,data):
    return (data.close[i]>data.open[i])

def stoploss(sl,trade_at,i,data):
    stop_at = trade_at*((100.0-sl)/100.0)
    return (stop_at>data.low[i] and stop_at<data.high[i])
```

# Task 1 

```python
def Task1(data,sl):
    bought = 0.0
    amount = 10000.0
    amt = []
    buy = []
    sell = []
    trade_at =0.0
    count = 1

    for i in range(1,len(data)):
        if(bought==0 and not(green(i,data)) and wick(i,data)>body(i,data) and data.volume[i]>data.volume[i-1]):
            bought = count
            amount -= count*data.high[i]
            trade_at = data.high[i]
            buy.append(i)
            print("trade at - ",i," - ",bought*data.high[i])
        elif(bought>0 and stoploss(sl,trade_at,i,data)):
            amount += bought*(trade_at*((100.0-sl)/100.0))
            print("stoploss sold at - ",i,"\n P/L = ",bought*(data.close[i]-(trade_at*((100.0-sl)/100.0))))
            bought = 0
            amt.append(amount)
            sell.append(i)
        elif(bought>0 and green(i,data) and wick(i,data)>body(i,data) and data.volume[i]>data.volume[i-1]):
            amount += bought*data.low[i]
            print("sold at - ",i," - ",(bought*data.low[i]),"\n P/L = ",bought*(data.low[i]-trade_at))
            bought = 0
            amt.append(amount)
            sell.append(i)
    
    if(bought>0):
        amount += bought*data.close[len(data)-1]
        print("EOD sold at - ",i,"\n P/L = ",bought*data.close[len(data)-1])
        bought = 0
        amt.append(amount)
        sell.append(i)
        
    return amount,amt,buy,sell
```
